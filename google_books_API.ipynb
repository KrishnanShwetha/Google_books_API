## Google Books API 

In this project I will be using data from the [google books API](https://developers.google.com/books/docs/overview)

Step 1) Create a list with these topic strings: Python; Data Science; Data Analysis; Machine Learning; and Deep 
Learning. Use these topics, one at a time, to query the Google Books API.

```python
import requests
import json
import pandas as pd

"""
    Google Books Api
    See: https://developers.google.com/books/
"""

def get(topic=""):
    BASEURL = 'https://www.googleapis.com/books/v1/volumes'
    headers = {'Content-Type': 'application/json'}

    response = requests.get(BASEURL + "?q=" + topic, headers=headers)

    if response.status_code == 200:
        return json.loads(response.content.decode('utf-8'))

    return response

topic_list = ['Python', 'Data Science', 'Data Analysis', 'Machine Learning', 'Deep Learning']
python = get('Python')
data_science = get('Data Science')
data_analytics = get('Data Analysis')
machine_learning = get('Machine Learning')
deep_learning = get('Deep Learning')
```

Step 2) b) For each returned JSON string: Convert the JSON string to a dict using `loads( )` then use this to convert it to a DataFrame: `pd.json_normalize( thedict['items'] )`. Then save them as `.csv`.

```python
# convert it into a dataframe 
# save it as a csv file

def read_file(filepath):
    df = pd.read_csv(filepath)
    return df


df1 = pd.json_normalize( python['items'])
csv_p = df1.to_csv("Python.csv")
#read_file("Python.csv")

df2 = pd.json_normalize(data_science['items'])
csv_dsc = df2.to_csv("data_science.csv")
#read_file("data_science.csv")

df3 = pd.json_normalize(data_analytics['items'])
csv_dan = df2.to_csv("data_analytics.csv")
#read_file("data_analytics.csv")

df4 = pd.json_normalize(machine_learning['items'])
csv_ml = df4.to_csv("machine_learning.csv")
#read_file("machine_learning.csv")

df5 = pd.json_normalize(deep_learning['items'])
csv_dl = df5.to_csv("deep_learning.csv")
read_file("deep_learning.csv")
```
